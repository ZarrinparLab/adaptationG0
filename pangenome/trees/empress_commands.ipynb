```python
import pandas as pd
```

```python
AZ51_aln = pd.read_csv('../AZ51_aln_names.txt', sep='\t', header=None)
AZ51_aln.rename(columns={0:'feature-id'}, inplace=True)
AZ51_aln[['mouse', 'timepoint', 'isolate', 'sample', 'bacteria', 'other']] = AZ51_aln['feature-id'].str.split('_', n=5, expand=True)
AZ51_aln.head()
```

```python
AZ51_aln['short-id'] = AZ51_aln['mouse'].astype(str)+"_"+AZ51_aln['timepoint'].astype(str)+"_"+AZ51_aln['isolate'].astype(str)
AZ51_aln.head()
```

```python
metadata = pd.read_csv('../../../metadata.txt', sep='\t')
metadata.rename(columns={'sample-id':'short-id'}, inplace=True)
metadata.head()
```

```python
fmd = pd.merge(AZ51_aln[['feature-id', 'short-id']], metadata, left_on='short-id', right_on='short-id')
fmd2 = fmd.set_index('feature-id')
fmd2.head()
```
