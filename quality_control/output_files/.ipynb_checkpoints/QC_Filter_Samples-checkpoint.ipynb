```python
import pandas as pd
import numpy as np
```

# AZ20

### load and prep files

```python
AZ20_checkmL = pd.read_csv("checkm_AZ20_lineage_summary.txt", sep='\t', index_col=0)
AZ20_checkmT = pd.read_csv("checkm_AZ20_taxonomy_summary.txt", sep='\t', index_col=0)
AZ20_quast = pd.read_csv("quast_AZ20_report.tsv", sep='\t', index_col=0)
AZ20_sourm = pd.read_csv("sourmash_AZ20_lca.csv")
```

```python
AZ20_checkmL.head()
```

```python
AZ20_checkmT.head()
```

```python
print(AZ20_checkmL.shape)
print(AZ20_checkmT.shape)
```

```python
AZ20_checkm = pd.merge(AZ20_checkmT, AZ20_checkmL, left_index=True, right_index=True)
AZ20_checkm.shape
```

```python
AZ20_checkm.head()
```

```python
AZ20_checkm.columns
```

```python
AZ20_quast.head()
```

```python
AZ20_quast.index
```

```python
AZ20_Quast = AZ20_quast.T
AZ20_Quast.head()
```

```python
AZ20_Quast['Genome fraction (%)'] = AZ20_Quast['Genome fraction (%)'].astype('float')
AZ20_Quast.dtypes
```

```python
AZ20_sourm[['other', 'sample_id']] = AZ20_sourm['ID'].str.split('/', expand=True)
AZ20_sourm.head()
```

```python
AZ20_sourm['sample'] = AZ20_sourm['sample_id'].replace(regex=['.fasta'], value='')
AZ20_sourm.head()
```

```python
AZ20_sourmash = AZ20_sourm.set_index('sample')
AZ20_sourmash.head()
```

```python
AZ20_sourmash.drop(columns=['species', 'strain', 'sample_id', 'other'], inplace=True)
AZ20_sourmash.shape
```

### merge

```python
AZ20_qc_part1 = pd.merge(AZ20_checkm, AZ20_Quast, left_index=True, right_index=True)
AZ20_qc_part1.head()
```

```python
AZ20_qc_all = pd.merge(AZ20_qc_part1, AZ20_sourmash, left_index=True, right_index=True)
AZ20_qc_all.head()
```

```python
AZ20_qc_all.shape
```

```python
AZ20_qc_all.to_csv('./AZ20_qc_all.csv')
```

### Filter

```python
AZ20_qc_all[(AZ20_qc_all['Contamination'] < 10)&(AZ20_qc_all['Strain heterogeneity'] != '100.0')
            &(AZ20_qc_all['Genome fraction (%)'] > 90)&(AZ20_qc_all['status'] == 'found')]
#dropped about 136 (11%), probably relax these parameters
```

```python
AZ20_qc_all[(AZ20_qc_all['Contamination'] > 10)|(AZ20_qc_all['Strain heterogeneity'] == '100.0')
            |(AZ20_qc_all['Genome fraction (%)'] < 90)|(AZ20_qc_all['status'] != 'found')]
#171 suspect samples - need to get that down to probably ~ 50
```

```python
AZ20_suspect = AZ20_qc_all[(AZ20_qc_all['Contamination'] > 10)|(AZ20_qc_all['Strain heterogeneity'] == '100.0')
            |(AZ20_qc_all['Genome fraction (%)'] < 90)|(AZ20_qc_all['status'] != 'found')]
AZ20_suspect.shape
```

```python
AZ20_suspect.to_csv('AZ20_suspects.txt', sep='\t')
```

```python
keep_list = pd.read_csv('../../AZ20_assembly_list.txt', sep='\t', header=None)
```

```python
red_list = pd.read_csv('AZ20_red_list.txt', sep='\t', header=None)
```

```python
Final_list = keep_list[~(keep_list[0].isin(red_list[0]))]
Final_list
```

```python
Final_list.to_csv('../../AZ20_passedqc_list.csv', header=None)
```
