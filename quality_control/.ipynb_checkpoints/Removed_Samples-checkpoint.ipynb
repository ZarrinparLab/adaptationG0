```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# AZ20

```python
Nissle_outliers = pd.read_csv('../AZ20_bad_genomes.txt')
Nissle_outliers.columns
```

```python
len(Nissle_outliers.columns)
```

```python
Nissle_qc = pd.read_csv('qc_output_files/AZ20_qc_all.csv')
Nissle_qc.rename(columns={'Unnamed: 0':'sample-id'}, inplace=True)
Nissle_qc.head()
```

```python
Nissle_qc['short-id'] = Nissle_qc['sample-id'].str.rsplit('_', n=1).str[0]
Nissle_qc
```

```python
bad_nissle = Nissle_qc[Nissle_qc['short-id'].isin(Nissle_outliers.columns)]
bad_nissle
```

```python
bad_nissle.to_csv('AZ20_removalset_completeinfo.csv')
```

# AZ51

```python
Native_outliers = pd.read_csv('../AZ51_bad_genomes.txt')
Native_outliers.columns
```

```python
len(Native_outliers.columns)
```

```python
Native_qc = pd.read_csv('qc_output_files/AZ51_qc_all.csv')
Native_qc.rename(columns={'Unnamed: 0':'sample-id'}, inplace=True)
Native_qc.head()
```

```python
Native_qc['long-sample-id'] = Native_qc.apply(lambda x: x['sample-id'].replace("_L002", ""), axis=1)
Native_qc['short-id'] = Native_qc['long-sample-id'].str.rsplit('_', n=1).str[0]
Native_qc
```

```python
bad_Native = Native_qc[Native_qc['short-id'].isin(Native_outliers.columns)]
bad_Native
```

```python
bad_Native.to_csv('AZ51_removalset_completeinfo.csv')
```
