```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# AZ20

```python
Nissle_allqc = pd.read_csv('qc_output_files/AZ20_qc_all.csv', index_col=0)
Nissle_allqc.head()
```

```python
Nissle_allqc2 = Nissle_allqc.reset_index()
Nissle_allqc2['sample-id'] = Nissle_allqc2['index'].str.rsplit('_', 1).str[0]
Nissle_allqc2.head()
```

```python
Nissle_allqc2.columns
```

```python
Nissle_primqc = Nissle_allqc2[['sample-id', 'Completeness', 'Contamination', 
                              '# contigs', '# contigs (>= 50000 bp)', 'Largest contig', 
                              'Total length', 'Reference length', 'Total aligned length', 
                              'Genome fraction (%)']].copy()
Nissle_primqc.head()
```

```python
Nissle_primqc.shape
```

```python
Nissle_plasmid_counts = pd.read_csv('../plasmids/AZ20_lplasmid_number.txt', sep='\t', header=None)
Nissle_plasmid_counts
```

```python
Nissle_plasmid_counts[['sample', 'other']] = Nissle_plasmid_counts[0].str.split('/', 1, expand=True)
Nissle_plasmid_counts['number_of_plasmids'] = Nissle_plasmid_counts['other'].str.split(':').str[1]
Nissle_plasmid_counts['sample-id'] = Nissle_plasmid_counts['sample'].replace(to_replace='_S[0-9]*', value='', regex=True)
Nissle_plasmid_counts['sample-id'] = Nissle_plasmid_counts['sample-id'].replace(to_replace='_L003', value='', regex=True)
Nissle_plasmid_counts['sample-id'] = Nissle_plasmid_counts['sample-id'].replace(to_replace='_pos', value='p', regex=True)
```

```python
for x in range(1,13):
    Nissle_plasmid_counts['sample-id'] = Nissle_plasmid_counts['sample-id'].replace(to_replace='_%s_2' % x, value='_%sx' % x, regex=True)
```

```python
Nissle_plasmid_counts['sample-id'] = Nissle_plasmid_counts['sample-id'].replace(to_replace='_12_A', value='_12', regex=True)
Nissle_plasmid_counts['sample-id'] = Nissle_plasmid_counts['sample-id'].replace(to_replace='_12_B', value='_13', regex=True)
```

```python
Nissle_plasmid_counts2 = Nissle_plasmid_counts[['sample-id', 'number_of_plasmids']].copy()
Nissle_plasmid_counts2.head()
```

```python
Nissle_qc_plas = pd.merge(Nissle_primqc, Nissle_plasmid_counts2, how='outer', 
                          left_on='sample-id', right_on='sample-id')
Nissle_qc_plas
```

```python
Nissle_removal = pd.read_csv('../AZ20_bad_genomes.txt')
Nissle_bad = Nissle_removal.columns
```

```python
Nissle_qc_plas = Nissle_qc_plas[~(Nissle_qc_plas['sample-id'].isin(Nissle_bad))].copy()
Nissle_qc_plas.shape
```

```python
Nissle_qc_plas['number_of_plasmids'] = Nissle_qc_plas['number_of_plasmids'].fillna(0)
#makes sure to account for everything that exists, even when it might not have detected plasmids
Nissle_qc_plas['number_of_plasmids'] = Nissle_qc_plas['number_of_plasmids'].astype(int)
#values treated as numbers
```

```python
Nissle_qc_plas['ideal_contig_num'] = Nissle_qc_plas['number_of_plasmids']+1
Nissle_qc_plas.head()
```

```python
Nissle_qc_plas.shape
```

```python
Nissle_qc_plas['Genome fraction (%)'].describe()
```

```python
Nissle_ideal = Nissle_qc_plas[Nissle_qc_plas['# contigs']<=Nissle_qc_plas['ideal_contig_num']]
Nissle_ideal.head()
```

```python
Nissle_ideal.shape
```

```python
Nissle_clean = Nissle_qc_plas[(Nissle_qc_plas['Completeness']>99)&(Nissle_qc_plas['# contigs (>= 50000 bp)']<2)&(Nissle_qc_plas['Largest contig']>4500000)]
Nissle_clean
```

```python
Nissle_clean.shape
```

```python
Nissle_best = Nissle_qc_plas[(Nissle_qc_plas['Genome fraction (%)']>99)]
#percentage of aligned bases in the reference genome. 
#A base in the reference genome is aligned if there is at least one contig with at least one alignment to this base.
Nissle_best
```
