```python
import pandas as pd
import numpy as np
```

```python
lplasmids = pd.read_csv('/Users/callaband/Downloads/AZ20_all_lplasmids.txt', sep='\t', header=None)
lplasmids[['random', 'sample-id', 'lplasmids']] = lplasmids[0].str.split("/", expand=True)
lplasmids
```

```python
gffs = pd.read_csv('/Users/callaband/Downloads/AZ20_all_current_gff.txt', sep='\t', header=None)
gffs[['random', 'sample-id', 'prokka', 'gff']] = gffs[0].str.split("/", expand=True)
gffs
```

```python
lpl = set(lplasmids['sample-id'].values)
gffl = set(gffs['sample-id'].values)
```

```python
missing = lpl.difference(gffl)
len(missing)
```

```python
df = pd.DataFrame(missing)
df
```

```python
df.to_csv('AZ20_missing.txt', sep='\t', header=None, index=False)
```
