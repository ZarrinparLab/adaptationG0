```python
lplasmids = pd.read_csv('/Users/callaband/Downloads/AZ51_all_lplasmids.txt', sep='\t', header=None)
lplasmids[['random', 'sample-id', 'lplasmids']] = lplasmids[0].str.split("/", expand=True)
lplasmids
```

```python
alf = pd.read_csv('/Users/callaband/Downloads/AZ51_list_le.txt', sep='\t', header=None)
alf[['mouse', 'timepoint', 'isolate', 'sample', 'lane']] = alf[0].str.split("_", expand=True)
alf
```

```python
alf['sample-id'] =alf['mouse'].astype(str)+"_"+alf['timepoint'].astype(str)+"_"+alf['isolate'].astype(str)+"_"+alf['sample'].astype(str)
```

```python
lpl = set(lplasmids['sample-id'].values)
alfl = set(alf['sample-id'].values)
```

```python
missing = alfl.difference(lpl)
len(missing)
```

```python
df = pd.DataFrame(missing)
df
```

```python
df.to_csv('AZ51_missing_prokka.txt', sep='\t', header=None, index=False)
```

```python
df51m = alf[alf['sample-id'].isin(list(missing))].copy()
df51m 
```
