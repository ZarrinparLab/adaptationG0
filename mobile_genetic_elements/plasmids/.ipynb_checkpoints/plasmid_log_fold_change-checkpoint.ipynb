# AZ20

```python
AZ20_genes = pd.read_csv('AZ20_gene_presence_absence_panaroo.Rtab', sep='\t', index_col=0)
AZ20_genes.head()
```

```python
AZ20_genes.columns = AZ20_genes.columns.str.replace("_AZ20_lplasmids", "")
AZ20_genes.columns = AZ20_genes.columns.str.replace("_AZ20_plasmids", "")
AZ20_genes.columns = AZ20_genes.columns.str.replace('_S[0-9]*', '', regex=True)
AZ20_genes.head()
```

```python
bytp = pd.merge(AZ20_genes.T, metadata[['collection_timepoint']], left_index=True, right_index=True)
bytp
```

```python
bytp.shape
```

```python
bytp_sum = bytp.groupby(by=['collection_timepoint']).sum()
```

```python
bytp_sum['isolate_count'] = bytp.groupby(by='collection_timepoint').count().relB.values
```

```python
columns_list = bytp.groupby(by=['collection_timepoint']).sum().columns
columns_list
```

```python
bytp_sum_prop = bytp_sum[columns_list].div(bytp_sum.isolate_count, axis=0)
bytp_sum_prop
```

```python
bytp_sum_prop_rev = bytp_sum_prop.T
bytp_sum_prop_rev_short = bytp_sum_prop_rev[['Day7', 'Month6']].copy()
bytp_sum_prop_rev_short
```

```python
bytp_sum_prop_rev_short['log2_fold_dif_M6_D7'] = np.log2(bytp_sum_prop_rev_short['Month6']) - np.log2(bytp_sum_prop_rev_short['Day7'])
```

```python
bytp_sum_prop_sort = bytp_sum_prop_rev_short.copy()
bytp_sum_prop_sort.replace([np.inf, -np.inf], np.nan, inplace=True)
bytp_sum_prop_sort.dropna(subset=['log2_fold_dif_M6_D7'], inplace=True)
bytp_sum_prop_sort
```

```python
bytp_sum_prop_sort.sort_values(by=['log2_fold_dif_M6_D7'])
```

```python
sns.scatterplot(data=bytp_sum_prop_sort, x='Day7', y='Month6')
```

```python
sns.ecdfplot(data=bytp_sum_prop_sort, x='log2_fold_dif_M6_D7')
```

```python
sns.histplot(data=bytp_sum_prop_sort, x='log2_fold_dif_M6_D7')
```

```python
bytp_sum_prop_sort_top = bytp_sum_prop_sort[(bytp_sum_prop_sort.log2_fold_dif_M6_D7>0.9)|(bytp_sum_prop_sort.log2_fold_dif_M6_D7<-2)]
bytp_sum_prop_sort_top.sort_values(by=['log2_fold_dif_M6_D7'])
```
