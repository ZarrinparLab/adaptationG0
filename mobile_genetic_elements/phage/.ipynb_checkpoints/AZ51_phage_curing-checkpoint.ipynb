## vMAGs 

```python
vMAG1 = pd.read_csv('AZ51_phage_query_vMAG1_reformat.txt', sep='\t', index_col=0)
vMAG1['query_id'] = 'vMAG-1'
vMAG1['sample_id'] = vMAG1['subject_acc'].str.split("_").str[0]
vMAG1['contig'] = vMAG1['subject_acc'].str.split("_").str[1]
vMAG1['contig'] = vMAG1['contig'].astype('int64')
vMAG1[['mouse', 'timepoint', 'isolate']] = vMAG1['sample_id'].str.split("-", n=2, expand=True)
vMAG1.head()
```

```python
vMAG1.shape
```

```python
vMAG1_nodups = vMAG1.drop_duplicates(subset=['sample_id'], keep='first')
vMAG1_filtered = vMAG1_nodups[(vMAG1_nodups['align_length']>5000) & (vMAG1_nodups['contig']<10)].copy()
# close to whole phage genome, well assembled isolate (also likely non-plasmid), 
vMAG1_filtered2 = vMAG1_filtered[vMAG1_filtered['subject_start'].between(1030000, 1040000, inclusive='both')]
#same location
vMAG1_filtered2
```

```python
vMAG2 = pd.read_csv('AZ51_phage_query_vMAG2_reformat.txt', sep='\t', index_col=0)
vMAG2['query_id'] = 'vMAG-2'
vMAG2['sample_id'] = vMAG2['subject_acc'].str.split("_").str[0]
vMAG2['contig'] = vMAG2['subject_acc'].str.split("_").str[1]
vMAG2['contig'] = vMAG2['contig'].astype('int64')
vMAG2[['mouse', 'timepoint', 'isolate']] = vMAG2['sample_id'].str.split("-", n=2, expand=True)
vMAG2.head()
```

```python
vMAG2.shape
```

```python
vMAG2_nodups = vMAG2.drop_duplicates(subset=['sample_id'], keep='first')
vMAG2_filtered = vMAG2_nodups[(vMAG2_nodups['align_length']>5000) & (vMAG2_nodups['contig']<10)].copy()
# close to whole phage genome, well assembled isolate (also likely non-plasmid), 
vMAG2_filtered2 = vMAG2_filtered[vMAG2_filtered['subject_start'].between(2400000, 2900000, inclusive='both')]
#same location
vMAG2_filtered2
```

```python
vMAG3 = pd.read_csv('AZ51_phage_query_vMAG3_reformat.txt', sep='\t', index_col=0)
vMAG3['query_id'] = 'vMAG-3'
vMAG3['sample_id'] = vMAG3['subject_acc'].str.split("_").str[0]
vMAG3['contig'] = vMAG3['subject_acc'].str.split("_").str[1]
vMAG3['contig'] = vMAG3['contig'].astype('int64')
vMAG3[['mouse', 'timepoint', 'isolate']] = vMAG3['sample_id'].str.split("-", n=2, expand=True)
vMAG3.head()
```

```python
vMAG3.shape
```

```python
vMAG3_nodups = vMAG3.drop_duplicates(subset=['sample_id'], keep='first')
vMAG3_filtered = vMAG3_nodups[(vMAG3_nodups['align_length']>5000) & (vMAG3_nodups['contig']<10)].copy()
# close to whole phage genome, well assembled isolate (also likely non-plasmid), 
vMAG3_filtered2 = vMAG3_filtered[vMAG3_filtered['subject_start'].between(3190000, 3210000, inclusive='both')]
#same location
vMAG3_filtered2
```

```python
allvMAGs = pd.concat([vMAG1_filtered2, vMAG2_filtered2, vMAG3_filtered2])
allvMAGs['timepoint'] = pd.Categorical(allvMAGs['timepoint'], ['Day3', 'Day7', 'Month1', 'Month3', 'Month6', 'Month10'])
allvMAGs
```

```python
sns.countplot(data=allvMAGs, x="query_id", hue="timepoint")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
sns.countplot(data=allvMAGs[allvMAGs['timepoint'].isin(['Day7', 'Month3', 'Month6'])], 
              x="query_id", hue="timepoint")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
together = pd.merge(allvMAGs, metadata, left_on='sample_id', right_index=True)
together['timepoint'] = pd.Categorical(together['timepoint'], ['Day3', 'Day7', 'Month1', 'Month3', 'Month6', 'Month10'])
```

```python
sns.catplot(data=together, x="query_id", hue="timepoint", kind='count', col='sex')
```

```python
together_filt = together[together['timepoint'].isin(['Day7', 'Month3', 'Month6'])].copy()
together_filt['timepoint'] = pd.Categorical(together_filt['timepoint'], [ 'Day7', 'Month3', 'Month6'])
```

```python
sns.catplot(data=together_filt,
            x="query_id", hue="timepoint", kind='count', col='sex')

plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'Arial'
plt.rcParams['svg.fonttype'] = 'none'
plt.savefig('figures/AZ51_filtered_vMAGs_counts.svg', dpi=150, bbox_inches='tight')
```
