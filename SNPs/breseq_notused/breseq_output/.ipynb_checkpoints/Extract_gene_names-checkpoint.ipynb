```python
import numpy as np
import pandas as pd
```

# AZ20

```python
AZ20_bindex = pd.read_csv('./AZ20_breseq_mutations.csv', index_col=0)
```

```python
AZ20_bindex.head()
```

```python
AZ20_bindex.shape
```

```python
AZ20_bindex['description']
```

```python
AZ20_bindex['gene_short'] = AZ20_bindex['description'].str.extract(r'\sGN=([a-zA-Z]+)\s',expand=False)
```

```python
AZ20_bindex['gene_short2'] = AZ20_bindex['description'].str.extract(r'\sGN=([a-zA-Z]+[0-9])\s',expand=False)
```

```python
AZ20_bindex['gene_organism'] = AZ20_bindex['description'].str.extract(r'\sOS=([a-zA-Z]+\s[a-zA-Z]+)\s',expand=False)
```

```python
AZ20_bindex
```

```python
additional = {"interleukin receptor mimic protein A": "irmA", 
              "hypothetical protein/interleukin receptor mimic protein A": "irmA", 
              "hypothetical protein": "hypothetical protein", 
             "hypothetical protein/hypothetical protein": "hypothetical protein", 
             'IS3 family transposase ISLad1': "ISLad1", 
             '16S ribosomal RNA (partial)': "16S rRNA (partial)", 
             'tRNA‑Ile(gat)/tRNA‑Ala(tgc)': "tRNA (gat or tgc)",
              '23S ribosomal RNA/5S ribosomal RNA': "23S/5S rRNA", 
             'Vitamin B12 transporter BtuB': 'BtuB', 
             'Outer membrane efflux protein':"OEP - tolC, yjcP (mdtP), yohG (mdtQ), or ylcB (cusC)",
             'Autotransporter beta‑domain protein': "agn43", 
             'O‑antigen polysaccharide polymerase Wzy': "Wzy (O antigen)", 
              'tRNA‑Gly(gcc)/hypothetical protein': "tRNA (gcc)", 
              'Terminase large subunit gpA, endonuclease domain': "gpA", 
              'D‑alanyl carrier protein': "dltC", 
              'D‑alanine‑‑D‑alanyl carrier protein ligase': "dltA", 
              'Outer membrane efflux protein/hypothetical protein': "OEP - tolC, yjcP (mdtP), yohG (mdtQ), or ylcB (cusC)", 
              '16S ribosomal RNA': "16S rRNA", 
              'tRNA‑Asn(gtt)': "tRNA (gtt)",
              'Sel1 repeat protein': "Sel1"}
```

```python
AZ20_bindex["additional_gene_info"] = AZ20_bindex["description"].map(additional)
```

```python
AZ20_bindex
```

```python
add_source = {"irmA" : "Escherichia coli", 
              "hypothetical protein": "unknown", 
             "ISLad1": "Escherichia coli"}
```

```python
AZ20_bindex["additional_gene_organism"] = AZ20_bindex["additional_gene_info"].map(add_source)
```

```python
AZ20_bindex2 = AZ20_bindex.assign(**{'combo_gene_info': AZ20_bindex['gene_short'].fillna(AZ20_bindex['gene_short2'])})
AZ20_bindex2 
```

```python
AZ20_bindex3 = AZ20_bindex2.assign(**{'all_gene_info': AZ20_bindex2['combo_gene_info'].fillna(AZ20_bindex2['additional_gene_info'])})
AZ20_bindex3 
```

```python
AZ20_bindex4 = AZ20_bindex3.assign(**{'organism_info': AZ20_bindex3['gene_organism'].fillna(AZ20_bindex3['additional_gene_organism'])})
AZ20_bindex4 
```

```python
AZ20_bindex5 = AZ20_bindex4.drop(columns=['combo_gene_info', 'additional_gene_info', 'gene_short', 
                                          'gene_short2', 'gene_organism', 'additional_gene_organism'])
AZ20_bindex5
```

```python
AZ20_bindex5.to_csv('./AZ20_breseq_mutations_mod.csv')
```

### Remove failed assemblies (did not pass qc)

```python
AZ20_keep_list = pd.read_csv('AZ20_passedqc_list.csv', usecols=[1], index_col=0, header=None)
AZ20_keep_list.shape
```

```python
AZ20_final = pd.merge(AZ20_bindex5, AZ20_keep_list, left_index=True, right_index=True)
AZ20_final.shape
```

```python
AZ20_final.to_csv('./AZ20_breseq_mutations_mod_qc.csv')
```
