```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gene_AZ20 = pd.read_csv("gene_results_AZ20.csv", index_col='Unnamed: 0')
pfam_AZ20 = pd.read_csv("pfam_results_AZ20.csv", index_col='Unnamed: 0')
gene_AZ51 = pd.read_csv("gene_results_AZ51.csv", index_col='Unnamed: 0')
pfam_AZ51 = pd.read_csv("pfam_results_AZ51.csv", index_col='Unnamed: 0')
```

```python
gene_AZ51.sort_values(by='pseudo_f', ascending=False)
```

```python
gene_AZ20.sort_values(by='pseudo_f', ascending=False)
```

# Common

```python
# commonalities between AZ51 and AZ20? 
# 3495 genes in common
common_genes = gene_AZ20.index.intersection(gene_AZ51.index)

gene_AZ51_common = gene_AZ51.loc[common_genes]
gene_AZ20_common = gene_AZ20.loc[common_genes]

# plot pseudo_fs against each other 
# make this look nicer
# essentially the takeaway here is that genes appear to change in one strain or another, not both 
sns.scatterplot(x=gene_AZ51_common.pseudo_f, y=gene_AZ20_common.pseudo_f)
```

```python
common_genes = pd.merge(gene_AZ20, gene_AZ51, left_index=True, right_index=True, suffixes=('_AZ20', '_AZ51'))
common_genes.head()
```

```python
common_genes.tail()
```

```python
def outlier_groups(x):
    if x < 10:
        return 'main'
    elif x >= 10:
        return 'outlier'
```

```python
common_genes['outlierAZ20'] = common_genes['pseudo_f_AZ20'].apply(outlier_groups)
common_genes['outlierAZ51'] = common_genes['pseudo_f_AZ51'].apply(outlier_groups)
common_genes
```

```python
common_genes['outlier'] = np.where(common_genes.outlierAZ20.str.contains('outlier'), 'outlier',
                       np.where(common_genes.outlierAZ51.str.contains('outlier'), 'outlier',
                       'main'))
```

```python
sns.scatterplot(data=common_genes, x='pseudo_f_AZ20', y='pseudo_f_AZ51', 
                hue='outlier', palette=['black', 'red'])
plt.axvline(10, linewidth=1, color='r', linestyle=':')
plt.axhline(10, linewidth=1, color='r', linestyle=':')
```

```python
common_genes[common_genes['outlier'] == 'outlier']
```

# Unique

```python
# do the genes that are unique to AZ20 have high pseudoFs 
unique_gene_AZ20 = gene_AZ20.loc[gene_AZ20.index.difference(common_genes)]
unique_gene_AZ51 = gene_AZ51.loc[gene_AZ51.index.difference(common_genes)]
```

```python
unique_gene_AZ20.sort_values(by='pseudo_f', ascending=False)
```
