```python
AZ20_filt1 = AZ20_raw.dropna(subset=['NT_POS']).copy()
AZ20_filt1.head()
```

```python
AZ20_filt1.shape
```

```python
AZ20_filt1.timepoint.value_counts()
```

```python
AZ20_filt1['sample_id'] = AZ20_filt1['mouse'].astype(str)+"_"+AZ20_filt1['timepoint'].astype(str)+"_"+AZ20_filt1['isolate'].astype(str)
AZ20_filt1.head()
```

```python
AZ20_filt1.shape
```

```python
AZ20_bad_genomes = pd.read_csv('../../AZ20_bad_genomes.txt')
AZ20_bad_genomes
```

```python
AZ20_filt1 = AZ20_filt1[~AZ20_filt1['sample_id'].isin(list(AZ20_bad_genomes.columns))]
AZ20_filt1.shape
```

```python
AZ20_filt1['syn_nonsyn'] = np.where(AZ20_filt1.EFFECT.str.contains("synonymous"), "synonymous",
                                       'non-synonymous')
AZ20_filt1.syn_nonsyn.value_counts()
```

```python
AZ20_filt1['all_mut'] = AZ20_filt1['TYPE']
        
for i in AZ20_filt1.index:
    if AZ20_filt1.at[i, 'TYPE'] == 'snp':
        if 'synonymous' in AZ20_filt1.at[i, 'EFFECT']:
            AZ20_filt1.at[i, 'all_mut'] = 'synonymous snp'
        else:
            AZ20_filt1.at[i, 'all_mut'] = 'non-synonymous snp'
    else:
        AZ20_filt1.at[i, 'all_mut'] = AZ20_filt1.at[i, 'TYPE']
```

```python
sns.countplot(data=AZ20_filt1, 
              x='timepoint', order=['Day7', 'Month3', 'Month6'], 
              hue='all_mut', hue_order=['mnp', 'ins', 'del', 'non-synonymous snp', 'complex', 'synonymous snp'],
             palette='winter')
plt.legend(loc='center left', bbox_to_anchor=(1.25, 0.5), ncol=1)
```

```python
AZ20_named = pd.merge(AZ20_filt1, metadata, left_on='sample_id', right_index=True)
AZ20_named.head()
```

```python
sns.countplot(data=AZ20_named, x='collection_timepoint', palette=['darkcyan', 'darkturquoise'],
              order=['Day7', 'Month3', 'Month6'], hue='syn_nonsyn')
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'Arial'
plt.rcParams['svg.fonttype'] = 'none'
plt.savefig('./AZ20_snps_rawcount_3tp.svg', dpi=150, bbox_inches="tight")
```

```python
AZ20_grouped = AZ20_named.groupby(['syn_nonsyn', 'collection_timepoint']).count()
AZ20_grouped2 = AZ20_grouped[['isolate']].reset_index().copy()
AZ20_grouped2['total_isolates'] = np.where(AZ20_grouped2.collection_timepoint.str.contains("Day3"), 71,
                                        np.where(AZ20_grouped2.collection_timepoint.str.contains("Day7"), 471,
                                                 np.where(AZ20_grouped2.collection_timepoint.str.contains("Month1"), 84,
                                                          np.where(AZ20_grouped2.collection_timepoint.str.contains("Month3"), 311,
                                                                   np.where(AZ20_grouped2.collection_timepoint.str.contains("Month6"), 235, 
                                                                            np.where(AZ20_grouped2.collection_timepoint.str.contains("Month10"), 49, np.nan))))))
AZ20_grouped2['count_per_total_isolates'] = AZ20_grouped2['isolate']/AZ20_grouped2['total_isolates']
AZ20_grouped2
```

```python
sns.catplot(data=AZ20_grouped2,
                  y='count_per_total_isolates', x='collection_timepoint', order=['Day7', 'Month3', 'Month6'],
                  hue='syn_nonsyn', hue_order=['synonymous', 'non-synonymous'],
                    kind='bar', palette=['darkcyan', 'darkturquoise'])
plt.ylim(0,4)
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'Arial'
plt.rcParams['svg.fonttype'] = 'none'
plt.savefig('./AZ20_snps_proportional_3tp.svg', dpi=150, bbox_inches="tight")
```
