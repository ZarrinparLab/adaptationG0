# AZ20

```python
AZ20_outliers = pd.read_csv('../pangenome/AZ20_outliers.txt')
AZ20_outliers.columns
```

```python
AZ20_metab = pd.read_csv('MA_AZ20_output/metabolic_summary__module_completeness.tsv', sep='\t', index_col=[2, 1, 0])
AZ20_metab.head()
```

```python
AZ20_metab.columns = AZ20_metab.columns.str.replace('_S([0-9]+)_AZ20_mut\.faa\.ko', '', regex=True)
AZ20_metab.columns = AZ20_metab.columns.str.replace('_S_AZ20_mut\.faa\.ko', '', regex=True)
AZ20_metab.head()
```

```python
AZ20_metab.shape
```

```python
AZ20_metab_nout = AZ20_metab.drop(columns=AZ20_outliers)
AZ20_metab_nout.shape
```

```python
AZ20_metab_nout.head()
```

```python
AZ20_metab_nout['median_completeness'] = AZ20_metab_nout.median(axis=1)
AZ20_metab_nout['mean_completeness'] = AZ20_metab_nout.mean(axis=1)
AZ20_metab_nout
```

```python
AZ20_metab_nout_high = AZ20_metab_nout[AZ20_metab_nout['median_completeness']>0].copy()
AZ20_metab_nout_high
```

```python
AZ20_ends_list = AZ20_metab_nout_high.columns[AZ20_metab_nout_high.columns.str.contains(r'Day7|Month6')]
AZ20_metab_nout_high_ends = AZ20_metab_nout_high[AZ20_ends_list].copy()
AZ20_metab_nout_high_ends.shape
```

```python
AZ20_ends_D7_list = AZ20_metab_nout_high_ends.columns[AZ20_metab_nout_high_ends.columns.str.contains(r'Day7')]
AZ20_ends_M6_list = AZ20_metab_nout_high_ends.columns[AZ20_metab_nout_high_ends.columns.str.contains(r'Month6')]
AZ20_metab_nout_high_ends['median_completeness_D7'] = AZ20_metab_nout_high_ends[AZ20_ends_D7_list].median(axis=1)
AZ20_metab_nout_high_ends['median_completeness_M6'] = AZ20_metab_nout_high_ends[AZ20_ends_M6_list].median(axis=1)
AZ20_metab_nout_high_ends['mean_completeness_D7'] = AZ20_metab_nout_high_ends[AZ20_ends_D7_list].median(axis=1)
AZ20_metab_nout_high_ends['mean_completeness_M6'] = AZ20_metab_nout_high_ends[AZ20_ends_M6_list].median(axis=1)
```

```python
AZ20_metab_nout_high_ends['change_median_completeness'] = (AZ20_metab_nout_high_ends['median_completeness_D7']-AZ20_metab_nout_high_ends['median_completeness_M6'])
AZ20_metab_nout_high_ends['change_mean_completeness'] = (AZ20_metab_nout_high_ends['mean_completeness_D7']-AZ20_metab_nout_high_ends['mean_completeness_M6'])
```

```python
AZ20_metab_nout_high_ends_median_change = AZ20_metab_nout_high_ends[AZ20_metab_nout_high_ends['change_median_completeness']>0]
AZ20_metab_nout_high_ends_median_change['trend'] = np.where(AZ20_metab_nout_high_ends_median_change['change_median_completeness'] > 0,   'decrease', 'increase')
AZ20_metab_nout_high_ends_median_change
#no median change in completeness
```

```python
AZ20_metab_nout_high_ends_mean_change = AZ20_metab_nout_high_ends[AZ20_metab_nout_high_ends['change_mean_completeness']>0]
AZ20_metab_nout_high_ends_mean_change['trend'] = np.where(AZ20_metab_nout_high_ends_mean_change['change_mean_completeness'] > 0,   'decrease', 'increase')
AZ20_metab_nout_high_ends_mean_change
#no mean change in completeness
```

# Plots by Pathways

```python
AZ51_colors=mcp.gen_color(cmap="autumn",n=33)
AZ51_colorz=random.sample(AZ51_colors, len(AZ51_colors))
AZ20_colors=mcp.gen_color(cmap="Blues",n=33)
AZ20_colorz=random.sample(AZ20_colors, len(AZ20_colors))
```

```python
AZ20_metab_nout_median_present_name = AZ20_metab_nout_median_present2.groupby('name')['median_completeness'].mean()
AZ20_metab_nout_median_present_name2 = pd.DataFrame(AZ20_metab_nout_median_present_name)
AZ20_metab_nout_median_present_name2.rename(columns={'median_completeness':'AZ20_median_completeness'}, inplace=True)
AZ20_metab_nout_median_present_name2.shape
```

```python
AZ20_metab_nout_median_present_name2
```

```python
AZ51_metab_nout_median_present_name = AZ51_metab_nout_median_present2.groupby('name')['median_completeness'].mean()
AZ51_metab_nout_median_present_name2 = pd.DataFrame(AZ51_metab_nout_median_present_name)
AZ51_metab_nout_median_present_name2.rename(columns={'median_completeness':'AZ51_median_completeness'}, inplace=True)
AZ51_metab_nout_median_present_name2.shape
```

```python
AZ51_metab_nout_median_present_name2
```

```python
together_completeness = pd.merge(AZ20_metab_nout_median_present_name2, AZ51_metab_nout_median_present_name2,
                     left_index=True, right_index=True, how='outer')
together_completeness['AZ20_median_completeness'] = together_completeness['AZ20_median_completeness'].fillna(0)
together_completeness['AZ51_median_completeness'] = together_completeness['AZ51_median_completeness'].fillna(0)
together_completeness
```

```python
together_completeness['difference'] = together_completeness['AZ20_median_completeness']-together_completeness['AZ51_median_completeness']
together_completeness
```

```python
together_differences = together_completeness[together_completeness['difference']!=0].copy()
together_differences
```

```python
together_differences['trend'] = np.where(together_differences['difference'] > 0,   'higher in AZ20', 'higher in AZ51')
together_differences
```

```python
together_differences['trend'].value_counts()
```

```python
AZ20_colors=mcp.gen_color(cmap="Blues",n=6)
AZ20_colorz=random.sample(AZ20_colors, len(AZ20_colors))
AZ51_colors=mcp.gen_color(cmap="Wistia",n=16)
AZ51_colorz=random.sample(AZ51_colors, len(AZ51_colors))
all_colorz = AZ20_colorz+AZ51_colorz
```

```python
together_differences['trend'] = pd.Categorical(together_differences['trend'], ['higher in AZ20', 'higher in AZ51'])
together_differences2 = together_differences.sort_values(by='trend')
together_differences2['colors'] = all_colorz
together_differences2
```

```python
together_differences2['abs_difference'] = np.abs(together_differences2['difference'])
```

```python
plt.figure(figsize=(15, 10), dpi=150)
plt.pie(together_differences2['abs_difference'].values,
        labels=together_differences2.index.values,
        colors=together_differences2.colors.values)

# add a circle at the center to transform it in a donut chart
my_circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)

plt.title('Pathway Differences', fontsize=18)
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'Arial'
plt.rcParams['svg.fonttype'] = 'none'
plt.savefig('./both_donut_differences.svg', dpi=300, bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(15, 10), dpi=150)
squarify.plot(sizes=together_differences2['abs_difference'].values, 
              label=together_differences2.index, 
              color=together_differences2.colors, alpha=.8 )
plt.axis('off')
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'Arial'
plt.rcParams['svg.fonttype'] = 'none'
plt.savefig('./both_treemap_bytrend.svg', dpi=300, bbox_inches="tight")
plt.show()
```
