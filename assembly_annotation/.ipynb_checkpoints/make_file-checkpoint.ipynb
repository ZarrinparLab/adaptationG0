```python
import numpy as np
```

```python
AZ20_datef['date'] = np.where(AZ20_datef[0].str.contains("ref"),"2017-07-12",
                   np.where(AZ20_datef[0].str.contains("Day3"), "2017-07-15",
                   np.where(AZ20_datef[0].str.contains("Day7"), "2017-07-19",
                   np.where(AZ20_datef[0].str.contains("Month1"), "2017-08-01",
                   np.where(AZ20_datef[0].str.contains("Month3"), "2017-10-17", 
                   np.where(AZ20_datef[0].str.contains("Month6"), "2018-01-16",
                   np.where(AZ20_datef[0].str.contains("Month10"), "2018-05-15", "other")))))))
```

```python
AZ20_datef['date'].unique()
```
